```python
import requests
import pandas as pd
from datetime import datetime
```

```python
api_key = "Api_Key"
baseurl = 'http://api.weatherapi.com/v1'
# f "http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}"
city = 'Louisville'
```

```python
response = requests.get(f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}")
response.status_code
```

```python
# Building a summary for the city and saving into a dictionary
# get a current weather and saving summary into a dictionary

def get_city_data(city, api_key):
    current_weather_url = f"{baseurl}/current.json?key={api_key}&q={city}"
    current_weather_response = requests.get(current_weather_url)
    if response.status_code == 200:
        data = current_weather_response.json()
        current = data['current']
        location = data['location']
        temp_f = current['temp_f']
        wind_mph = current['wind_mph']
        humidity = current['humidity']

        print('Summary of the City:')
        print(f" Weather for {location['name']}, {location['region']} ")
        print(f" Temperature: {current['temp_f']} °F /  {current['temp_c']}  °C")
        print(f" Condition: {current['condition']['text']}")
        print(f" Humidity: {current['humidity']}%")
        print(f" Wind Speed {current['wind_mph']} mph / {current['wind_kph']} kph")
        print(f" Last Update: {current['last_updated']}")
        city_summary_dictionary = {
            'city_name': location['name'],
            'region': location['region'],
            'country': location['country'],
            'temperature_c': current['temp_c'],
            'temperature_f': current['temp_f'],
            'condition': current ['condition']['text'],
            'humidity': current['humidity'],
            'wind_kph': current['wind_kph'],
            'last_updated': location['localtime']}
    
        return city_summary_dictionary

    else:
        print(f"Error fetching data for {city}:{response.status_code}")

        

summary = get_city_data(city, api_key)


print(summary)
        
```

```python
# Get a forecast for 7 days
def get_forecast(city, api_key):
    forecast_url= f"{baseurl}/forecast.json?key={api_key}&q={city}&days=7"
    forecast_response = requests.get(forecast_url)
    forecast_data = forecast_response.json()
    forecast = forecast_data['forecast']['forecastday']
    forecast_7_days = []
    for day in forecast:
        date_obj = datetime.strptime(day['date'], '%Y-%m-%d')
        formatted_date = date_obj.strftime('%A, %B %d')
        forecast_7_days.append(f"{formatted_date}: High {day['day']['maxtemp_f']}F, Low{day['day']['mintemp_f']}F - {day['day']['condition']['text']}")
    return forecast_7_days

result = get_forecast(city, api_key)

for day in result:
    print(day)
```

```python
#get some astronomical data

def get_astronomy_data(city, api_key):
    today = pd.Timestamp.now().strftime("%Y-%m-%d")
    response = requests.get(f"{baseurl}/astronomy.json?key={api_key}&q={city}&dt={today}")
    astronomy_data = response.json()
    astro = astronomy_data['astronomy']['astro']
    print(f"Sunrise: {astro['sunrise']}")
    print(f"Sunset: {astro['sunset']}")
    print(f"Moonrise: {astro['moonrise']}")
    print(f"Moonset: {astro['moonset']}")  
    print(f"Moon Phase: {astro['moon_phase']}") 

prediction = get_astronomy_data(city, api_key)
```
